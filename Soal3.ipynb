```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
df = pd.read_excel('Dataset UAS.xlsx', sheet_name='emoney pilihan')
# Jumlah responden
n = 891
# confidence level
c_level = 0.99

display(df)
```

```python
# Selang kepercayaan untuk p1 - p2
z_score = st.norm.ppf((1+c_level)/2)
em1 = "GoPay"
em2 = "ShopeePay"

n1 = df.loc[df['Jenis eMoney'] == em1, 'Jumlah responden yang menggunakan'].values[0]
n2 = df.loc[df['Jenis eMoney'] == em2, 'Jumlah responden yang menggunakan'].values[0]
p1 = n1 / n
p2 = n2 / n

# n1 = n
# n2 = n
# p1 = 0.47
# p2 = 0.153
p_diff = p1 - p2

error = z_score * np.sqrt((p1 * (1 - p1) / n1) + (p2 * (1 - p2) / n2))
confidence_interval = (p_diff - error, p_diff + error)
print(f"Confidence Interval: {confidence_interval}")
```
